# Importing the libararies

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
```

# importing the dataset


```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[ : , [3 , 4] ].values
```

# using the elbow method to find the optimal no of clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmean = KMeans(n_clusters = i , init = 'k-means++' , random_state = 42 ) 
    kmean.fit(x)
    wcss.append(kmean.inertia_)
plt.plot(range(1,11) ,wcss)
plt.title('the elbow method')
plt.xlabel(' no of clusters' )
plt.ylabel('wcss')
plt.show()
```

# Training the k mean model

```python
kmean = KMeans(n_clusters = 5 , init = 'k-means++' , random_state = 42 ) 
ykmean = kmean.fit_predict(x)
```

```python
print(ykmean)
```

## Visulaizng the clusters

```python
plt.scatter(x[ykmean == 0 , 0 ] , x[ykmean == 0 , 1] , s = 100 , color = 'red' , label = 'cluster1')
plt.scatter(x[ykmean == 1 , 0 ] , x[ykmean == 1 , 1] , s = 100 , color = 'yellow' , label = 'cluster2')
plt.scatter(x[ykmean == 2 , 0 ] , x[ykmean == 2 , 1] , s = 100 , color = 'green' , label = 'cluster3')
plt.scatter(x[ykmean == 3 , 0 ] , x[ykmean == 3 , 1] , s = 100 , color = 'blue' , label = 'cluster4')
plt.scatter(x[ykmean == 4 , 0 ] , x[ykmean == 4 , 1] , s = 100 , color = 'pink' , label = 'cluster5')
plt.scatter(kmean.cluster_centers_[ : , 0 ] , kmean.cluster_centers_[ : , 1 ] , c = 'cyan' , s = 300 , label = 'centroid')
plt.title('clusters of clusters')
plt.xlabel('annual income ')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
